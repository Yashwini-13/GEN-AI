```python
dt = dt.drop('Gender',axis = 1)
```

```python
x = dt.drop('Salary',axis = 1)
y = dt['Salary']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.25)
```

splitting the data because we only have one set of data...so split it and train using half of data and test using other data... size is 0.25 beacuse 25% for testing and balance for training

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
```

using Linear regression since the graph of education in salary is linear....we use this..based on the graph we choose our algorithm
y = mx +c
y = m1 x education+ m2 x job..............etc

```python
y_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

using r2_score to chk the dissimilarities

```python
c = lr.intercept_
m = lr.coef_
m1,m2,m3,m4 = m[0],m[1],m[2],m[3]
```

m is slope

```python
age = 35
el = 1
jt = 150
exp = 8
salary = m1*age + m2*el + m3*jt + m4*exp
print('Salary',salary)
```
