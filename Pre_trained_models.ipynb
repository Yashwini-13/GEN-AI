```python
from textblob import TextBlob
#text = 'The service was terrible & food was mediocre'
#text = "The weather is quite pleasant today"
text = "i broke my leg"
blob = TextBlob(text)
p = blob.sentiment.polarity
print("Positive" if p>0 else "Negative" if p<0 else "Neutral")
```

do not give any super adjective word it shows neutral only.....................TextBlob uses a lexicon-based sentiment analyzer.

It looks for words in its dictionary (like happy, sad, hate, love) that have polarity scores.

In your text "i broke my leg", none of the words (i, broke, my, leg) have sentiment scores in the lexicon........................we can use flair or spacywhich gives crt output

```python
import spacy
nlp = spacy.load('en_core_web_sm')
text = '''Last summer, Rahul Sharma traveled from Delhi to New York for a business conference.
He stayed there for two weeks and met with Dr. Emily Johnson at Columbia University.
 After that, he spent three days in Boston before flying back to India on August 15, 2023.'''
doc = nlp(text)
for ent in doc.ents:
  print(ent.text,ent.label_)
```

```python
from spacy import displacy
displacy.render(doc,style='ent',jupyter=True)
```
