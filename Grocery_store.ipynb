```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

apriori - used for collecting frequent bought it(algorithm)

read row by row and remove the null value of each row and creating the list

```python
dt = pd.read_csv('/content/groceries.csv',header=None)
dt.head()
```

since there is no heading in the data give header is none else it will take the first row as header

```python
trnx = dt.apply(lambda x: x.dropna().tolist(), axis=1).tolist()
trnx[0:4]
```

transaction for removing null values row by row..apply is used(two dimensional list )

```python
te = TransactionEncoder()
te_ary = te.fit(trnx).transform(trnx)
df = pd.DataFrame(te_ary, columns=te.columns_)
df.head()
```

transactional encoding....making all the products as y column and making t or f for each product if the customer has brought

again analyse this data and separate frequently bought products(called as frequent item extraction)

```python
freq_items = apriori(df, min_support=0.03, use_colnames=True)
freq_items.tail()
```

apriori algorthim is used

after frequent brought together...next step association rule...........if some product is choosed the other products can also be recommended

```python
rules = association_rules(freq_items, metric="confidence", min_threshold=0.03)
rules.head()
```

antecendts the product they choose and consequents are product recommended

```python
print(rules.iloc[:,[0,1]])
```

```python
print(rules[['antecedents','consequents','lift']].sort_values(by = 'lift',ascending=False).head(10))
```

if the product chosen by customer is not in antecedents it will not recommend any other products....

```python
product = input("Enter a product:").lower()
prd_rule = rules[rules['antecedents'].apply(lambda x: product in x)]
if not prd_rule.empty:
  print(f"since you brought {product},you would also like:")
  for i in prd_rule['consequents']:
    print(list(i)[0])
else:
    print("No recommendation found")
```
